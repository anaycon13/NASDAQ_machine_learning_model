```python
import pandas as pd
#VIF method for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor 
import numpy as np

path = '/content/drive/Shareddrives/MATH456'

data = pd.read_csv(path + '/Cleaned.csv')

#just want the independent columns from here
data = data.drop(['quarter','NASDAQCOM'], axis=1)

def calc_vif(X):
  vif = pd.DataFrame()
  vif["variables (quarterly)"] = X.columns
  vif["VIF"] = [np.round(variance_inflation_factor(X.values, i),2) for i in range(X.shape[1])]

  return vif

#Visualizing the correlation matrix to help us access which data we want to keep
#and remove when we use model implementation
np.round(data.corr(),2)
```

```python
#Using variance inflation factor to help access which variables to keep and drop
#for the model of choice
calc_vif(data)
```

```python
#Using the table from above, we will drop the variables with the highest VIFs
new_data = data.drop(['Consumer Price Index for All Urban Wage Earners and Clerical Workers: ' + 
                      'Information Technology, Hardware and Services in U.S. City Average',
                      'Large Bank Consumer Credit Card Balances: Total Balances'], axis=1)

calc_vif(new_data)
```
